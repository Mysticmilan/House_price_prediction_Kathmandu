```python
dataset.drop(dataset[(dataset['District'] == 'Lalitpur') | (dataset['District'] == 'Bhaktapur')].index, inplace=True)
dataset
```

```python
def convert_months_to_years(months):
    if months <= 12:
        return 1
    elif 12 < months <= 24:
        return 2
    elif 24 < months <= 36:
        return 3
    elif 36 < months <= 48:
        return 4
    elif 48 < months <= 60:
        return 5
    elif 60 < months <= 72:
        return 6
    else:
        return months // 12

dataset['Posted Date(Moth Ago)'] = dataset['Posted Date(Moth Ago)'].apply(convert_months_to_years)
```

```python
dataset.drop(['Build Year(In B.S.)', 'District'], axis = 1, inplace = True)
```

```python
# Custom function to fill null and zero values
def fill_null_zero(data, column, value):
    data[column] = data[column].fillna(value).replace(0, value)

# Fill null and zero values with specified values
fill_null_zero(dataset, 'Bathroom', 3)
fill_null_zero(dataset, 'Parking', 2)
fill_null_zero(dataset, 'Road Distance(Feet)', 14)
fill_null_zero(dataset, 'Type of Road', 'Gravelled')
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```

```python
dataset['Property Face'] = label_encoder.fit_transform(dataset['Property Face'])
```

```python
dataset['Type of Road'] = label_encoder.fit_transform(dataset['Type of Road'])
```

```python
dataset['Location'] = label_encoder.fit_transform(dataset['Location'])
```

```python
dataset.drop(['Property Face','Parking','Location','Type of Road'], axis = 1, inplace = True)
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = dataset.drop('Price(NRS.)', axis=1)
y = dataset['Price(NRS.)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predecitions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predecitions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test, rfc_pred))
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
